```python
import re
import nltk

nltk.download('stopwords')
nltk.download('punkt') 
nltk.download('wordnet') 

from nltk.corpus import stopwords
```

Performing text processing: for 'title' column in the 'messages' dataFrame by removing non-alphanumeric characters, converting to lowercase, splitting into words, applying stemming, removing stopwords, and then joining the processed words back into a string. The resulting strings are stored in a list called 'corpus'.

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z0-9]',' ', str(messages['title'][i]))
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

Text preprocessing pipeline using the NLTK library

The below code Iterates through the 'content' column of the 'messages' DataFrame, cleans each text entry by removing non-alphanumeric characters, converting to lowercase, and lemmatizing words while excluding stopwords:

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

cleaned_corpus = []
for i in range(len(messages)):
    review = re.sub('[^a-zA-Z0-9]',' ',str(messages['content'][i]))
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    cleaned_corpus.append(review)
```

Removing empty sentences from the corpus:

```python
cleaned_corpus_ = [i for i in cleaned_corpus if i]
len(cleaned_corpus_)
```

Similarly Dropping Those rows in Y w.r.t Corpus which were empty Strings

```python
Y = df[list(map(lambda x: len(x)>0,cleaned_corpus))]
Y = Y['label']
```

Doing one-hot encoding:

```python
voc_size=5000
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
onehot_repr=[one_hot(words,voc_size)for words in cleaned_corpus_]
onehot_repr
```

```python
cleaned_corpus_[2]
```

```python
# Finding the Max_len & Min_len of the Sentences w.r.t Cleaned_Corpus

def max_minlen(cleaned_corpus_):
  max_l = float('-inf')
  min_l = float('inf')
  for i in cleaned_corpus_:
    if len(i.split())>max_l:
      max_l = len(i.split())
    if len(i.split())<min_l:
      min_l = len(i.split())
  return max_l,min_l # max_l & min_l
max_minlen(cleaned_corpus_)
```

```python
onehot_repr[2]
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# setting max sent length: 
sent_length=40 
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length) # Pre - adding zeros at the beg
print(embedded_docs)
```

```python
embedded_docs[1]
```

```python
embedded_docs[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
# setting max sent length
embedding_vector_features=50 
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))

# setting 100 Neurons for Hyper-Parameter for LSTM
model.add(LSTM(100)) 

# Output Layer
model.add(Dense(1,activation='sigmoid')) 
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy']) # Compiling the model
```

```python
print(model.summary())
```

```python
len(embedded_docs),Y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
Y_final=np.array(Y)
```

```python
X_final.shape,Y_final.shape
```

```python
# Train & Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, Y_final, test_size=0.30, random_state=0)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=5,batch_size=20)
```

```python
import tensorflow as tf
callback = tf.keras.callbacks.EarlyStopping(patience=4, mode='min', restore_best_weights=True) # Early Stopping
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=5,batch_size=20,callbacks=[callback])
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.5, 1,0)
```

```python
if (y_pred[0]==0):
  print(f'The News is Realiable & labelled as {y_pred[0]} and in real it is {y_test[0]}')
else:
  print(f'The News is Fake & labelled as {y_pred[0]} and in real it is {[y_test[0]]}')
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

Accuracy achieved using LSTM: 98%
